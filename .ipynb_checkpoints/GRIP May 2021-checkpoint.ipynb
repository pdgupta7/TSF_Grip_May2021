# Exploratory Data Analysis

```python
data.describe()
```

```python
#Calculating Total Sales, Profit and Quantity
print("Total Sales {}".format(sum(data['Sales']),2))
print("Total Profit {}".format(sum(data['Profit']),3))
print("Total Qunatity {}".format(sum(data['Quantity']),3))
```

Getting into more details of quantity. Trying to dig in more with respect to features Category, Segment, Region and Ship Mode

Feature: Category

```python
print("Feature: Category")
print("Below graph shows number of transaction with respect to Category: Furniture, Office Supplies Category and Technology")
sns.countplot(data['Category'], palette = 'mako')
plt.tight_layout()
```

```python
print("Feature - Category")
data_category = data["Category"].count()
print("Total Records",data_category)
data_category_sub = data["Category"].value_counts()
print(data_category_sub)
```

```python
data_category_1 = data.groupby("Category")[["Profit","Sales"]].sum()
print(data_category_1)
data_category_1.plot(kind="bar", figsize=(15,5), title="Category: Sales and Profit", fontsize = 10, colormap = "Dark2")
```

1. Most of the transactions are Recorded for Category: Office Supplies and least for Category: Technology
2. On the Other hand, even after having the least number of transactions for Technology, the transaction value for Technology is highest which marks as maximum sales value and thus achieving the maximum profit. Whereas, Category: Furniture has the lowest profit.
3. Maximum Sales value achieved by Category:Technology whereas Lowest for Category:Office Supplies

Feature: Sub-Category

```python
print("Feature - Sub-Category")
data_subcategory = data["Sub-Category"].count()
print("Total Records",data_subcategory)
data_subcategory_ = data["Sub-Category"].value_counts()
print(data_subcategory_)
```

```python
data_subcategory_1 = data.groupby("Sub-Category")[["Profit","Sales"]].sum()
print(data_subcategory_1)
data_subcategory_1.plot(kind="barh", figsize=(15,5), title="Sub-Category: Sales and Profit", fontsize = 10, colormap = "winter")
```

For Feature: Sub-Category
1. Maximum Sales Transactions are recorded for Binders(1518) and Papers(1314)
2. Maximum Sales Value is achieved by Phones(329704.67) and Chairs(327777.76)
3. In Terms of Profit, Copiers have marked the maximum profit followed by Phones(44493.05) and Accessories(41920.90).
4. Sub-Category: Tables have incurred major loss of -17725.48

To understand detailed relationship between Sales and Profit we are using Scatterplot

```python
fig, ax = plt.subplots(figsize = (12, 8))
ax.scatter(data["Sales"] , data["Profit"], color = 'Teal')
ax.set_xlabel('Sales', fontsize=15)
ax.set_ylabel('Profit | Loss', fontsize=15)
plt.title('Sales and Profit Relationship', fontsize = 25)
plt.show()
```

Conclusion from above Relationship Scatterplot 

1. Most of the Sales are under 5000
2. Majority of Sales Above 5000 results in Profit for the Business
3. Huge number of Sales (Transactions) between 0 to 2500 results in Loss

Feature: State

```python
print("Feature - State")
data_state = data["State"].count()
print("Total Records",data_state)
data_state_ = data["State"].value_counts()
print(data_state_)
```

```python
data_state_transact = data.groupby('State')['Quantity'].count().sort_values(ascending = False).plot.bar(figsize = (20, 5), color = 'Teal')
plt.ylabel('Total Number of Transactions')
plt.xlabel('States')
plt.title('Total State Wise Transactions', fontsize = 15)
plt.show()
```

1. Maximum Transactions in California
2. Second highest transactions in New York (way behind California)
3. Third highest transactions is Texas
4. Lowest transactions recorded in states North Dakota, West Virginia and least in Wyoming

Total Sales in Each State

```python
data_state_sales = data.groupby('State')['Sales'].sum().sort_values(ascending = False).plot.bar(figsize = (20, 5), color = 'Teal')
plt.ylabel('Total Sales')
plt.xlabel('States')
plt.title('State-Wise TotalSales', fontsize = 15)
plt.show()
```

1. Again, California tops the chart with maximum number of sales followe by New York and Texas.
2. Whereas, Maine, West Virginia, North Dakota are the ones with least number of sales.

Profit & Loss for each State

```python
data_state_profit_loss = data.groupby('State')['Profit'].sum().sort_values(ascending = False).plot.bar(figsize = (20, 5), 
                                                                                              color = 'Green')
plt.ylabel('Profit | Loss')
plt.xlabel('States')
plt.title('State-Wise Profit & Loss', fontsize = 15)
plt.show()
```

```python
data_state_discount = data.groupby('State')['Discount'].mean().sort_values(ascending = False).plot.bar(figsize = (20, 5), color = 'Teal')
plt.ylabel('Average Discount')
plt.xlabel('States')
plt.title('State Wise Discount', fontsize = 15)
plt.show()
```

From above graphs we can conclude:
1. For Texas, Illinois, Pennsylvania even after with the highest discounts these states have sufferred a great loss. This confirms the Negative Correlation (-0.22) above As Discount increases Profit Decreases.
2. California with the highest profit stands on 14th position getting Average discount of 0.07% whereas New York with second highest profit stands on 20th position with average discount of 0.05%. Also, California and New York mark as states with maximum number of sales transactions, therefore it confirms the Positive Correlation (0.48) between Sales and Profit.
3. Majority of the states are marked with a small or no profit with Average Loss of 83% which is huge

Profit and Discount Relationship

```python
sns.lineplot(x="Discount",y="Profit", data=data)
plt.show()
```

Above line graph confirms the negative relationship between the Profit and the Discount.
As Discount increases Profit decreases.
